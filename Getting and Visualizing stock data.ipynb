```python
# This code explores the yfinance library to get stock data from yahoo finance, and altair to visualize the data.

# Packages requires:
    # - pandas    >>>  pip install pandas    >>> library for converting data into DataFrames & save in csv or xlxs format
    # - yfinance  >>>  pip install yfinance  >>> package that will allow to access stock data
    # - altair    >>>  pip install altair    >>> statistical visualization library

# Steps:
    # - import required libraries 
    # - define a function with one arg and plot chart inside the function
    # - call function by passing company's name and chart will display
```

```python
import yfinance as yf
import altair as alt
import pandas as pd
```

```python
def get_data(company_name):
    data = yf.download(company_name, period='7y')
    price = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Close',
    tooltip=['Date', 'Open', 'High', 'Low', 'Volume']).properties(
    width=800, height=300)
    
    volume = alt.Chart(data.reset_index()).mark_line().encode(
    x='Date:T',
    y='Volume').properties(width=800, height=60)
       
    return price & volume
```

```python
get_data('GOOGL')
```
